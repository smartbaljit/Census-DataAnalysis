```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
```

```python
csv_path = "Resources/NJ_CensusGeoDataEssexCounty.csv"

# Read the CSV into a Pandas DataFrame
census_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
census_df.head()
```

```python
# Count the total rows in each column
census_df.count()
```

```python
# Round the Average Price and Average Ratings columns
census_df=census_df.round(2)
census_df
```

```python
plt.scatter(census_df["Median Age"], census_df["Average Price"], marker="o", facecolors="blue", edgecolors="black")
plt.ylabel("Average Price")
plt.xlabel("Median Age")
plt.title(f"Median Age Vs Average Price")
plt.grid()
plt.savefig("Images/Age_AvgPrice.png")
plt.show()
```

```python
plt.scatter(census_df["Median Age"], census_df["Average Rating"], marker="o", facecolors="blue", edgecolors="black")
plt.ylabel("Average Rating")
plt.xlabel("Median Age")
plt.title(f"Median Age Vs Average Rating")
plt.grid()
plt.savefig("Images/Age_AvgRating.png")
plt.show()
```

```python
plt.scatter(census_df["Foreign Born Percent"], census_df["Average Price"], marker="o", facecolors="blue", edgecolors="black")
plt.ylabel("Average Price")
plt.xlabel("Foreign Born Percent")
plt.title(f"Foreign Born Percent Vs Average Price")
plt.grid()
plt.savefig("Images/ForBornPerc_AvgPrice.png")
plt.show()
```

```python
plt.scatter(census_df["Foreign Born Percent"], census_df["Average Rating"], marker="o", facecolors="blue", edgecolors="black")
plt.ylabel("Average Rating")
plt.xlabel("Foreign Born Percent")
plt.title(f"Foreign Born Percent Vs Average Rating")
plt.grid()
plt.savefig("Images/ForBornPerc_AvgRating.png")
plt.show()
```

```python
plt.scatter(census_df["Household Income"], census_df["Average Price"], marker="o", facecolors="blue", edgecolors="black")
plt.ylabel("Average Price")
plt.xlabel("Household Income")
plt.title(f"Household Income Vs Average Price")
plt.grid()
plt.savefig("Images/HouseholdIncome_AvgPrice.png")
plt.show()
```

```python
plt.scatter(census_df["Household Income"], census_df["Average Rating"], marker="o", facecolors="blue", edgecolors="black")
plt.ylabel("Average Rating")
plt.xlabel("Household Income")
plt.title(f"Household Income Vs Average Rating")
plt.grid()
plt.savefig("Images/HouseholdIncome_AvgRating.png")
plt.show()
```
