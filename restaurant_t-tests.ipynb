```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint
import scipy.stats as stats
```

```python
# These files are a combo of census api data and google places api data.
# see census_api.ipynb and google_api.ipynb for how data was pulled

nj_age_young_csv = "Resources/NJ_Age_Young_Data.csv"
nj_age_old_csv = "Resources/NJ_Age_Old_Data.csv"
nj_income_low_csv = "Resources/NJ_Income_Poor_Data.csv"
nj_income_high_csv = "Resources/NJ_Income_Rich_Data.csv"
nj_foreign_low_csv = "Resources/NJ_Foreign_Low_Data.csv"
nj_foreign_high_csv = "Resources/NJ_Foreign_High_Data.csv"
```

```python
# read all the csv
nj_age_young_df = pd.read_csv(nj_age_young_csv, index_col=0)
nj_age_old_df = pd.read_csv(nj_age_old_csv, index_col=0)
nj_income_low_df = pd.read_csv(nj_income_low_csv, index_col=0)
nj_income_high_df = pd.read_csv(nj_income_high_csv, index_col=0)
nj_foreign_low_df = pd.read_csv(nj_foreign_low_csv, index_col=0)
nj_foreign_high_df = pd.read_csv(nj_foreign_high_csv, index_col=0)
```

```python
# each dataframe is a sample population of 25 zip codes
# create a list with 25 numbers to act as our x axis

samples_list = list(range(25))
```

```python
# plot average price of restaurants in Top 25 youngest and oldest nj zip codes

plt.figure(figsize=(12, 8))

young_plot = plt.scatter(
    samples_list, nj_age_young_df["Average Price"],
    marker = "o", facecolor="blue", edgecolor="black", label="Young Zip Codes")

old_plot = plt.scatter(
    samples_list, nj_age_old_df["Average Price"],
    marker = "x", facecolor="orange", edgecolor="black", label="Old Zip Codes")

# set labels and graph features
plt.title(f"Average Restaurant Price of 25 Youngest and Oldest Zip Codes in NJ", fontsize=16)

plt.xlim(-1, 25)
plt.xlabel("Zip Code Samples", fontsize=16)

plt.ylabel("Average Price (Google Code)", fontsize=16)

plt.legend()

plt.grid()

# save graph image
save_age_price = "Images/NJ_Median_Age_Price_25_Plot"
plt.savefig(save_age_price, bbox_inches="tight")

plt.show()
```

```python
# Do a t-test on the two populations of prices

age_price_ttest = stats.ttest_ind(nj_age_young_df["Average Price"], nj_age_old_df["Average Price"], equal_var=False)
age_price_pvalue = age_price_ttest[1]

print(f"P-Value is {age_price_pvalue:,.16f}")

if age_price_pvalue < 0.05:
    print("Difference in restaurant pricing for different age groups IS statisically significant")
else:
    print("Difference in restaurant pricing for different age groups is NOT statisically significant")
```

```python
# build a dataframe of p-values

demo_data = ["Median Age", "Household Income", "Foreign Born"]
price_data = [age_price_pvalue, income_price_pvalue, foreign_price_pvalue]
rating_data = [age_rating_pvalue, income_rating_pvalue, foreign_rating_pvalue]

ttest_data = zip(demo_data, price_data, rating_data)
ttest_df = pd.DataFrame(ttest_data, columns=["Demographic", "Price pvalue", "Rating pvalue"])

ttest_df["Price pvalue"] = ttest_df["Price pvalue"].map("{0:,.16f}".format)
ttest_df["Rating pvalue"] = ttest_df["Rating pvalue"].map("{0:,.16f}".format)

ttest_df = ttest_df.set_index("Demographic")
ttest_df
```
