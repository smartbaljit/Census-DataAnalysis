```python
# Dependencies and Setup
#%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import tkinter as tk

# File to Load (Remember to Change These)
NJ_Census_Geo_Data_Demo = "Census_Google_Essex.csv"

# Read Census Data File and store into Pandas Data Frames
nj_census_data_df = pd.read_csv(NJ_Census_Geo_Data_Demo)

nj_census_data_df.head()
```

```python
#Extract Essex County Data
Essex_df = nj_census_data_df[nj_census_data_df['County']=='Essex']
Essex_df.head()
```

```python
hh_income = []
```

```python
#Calculate Household Income
hh_income = Essex_df.groupby(['City'])['Household Income'].mean()
#hh_income.head()
```

```python
#Calculate Median Age
med_age = Essex_df.groupby(['City'])['Median Age'].mean()
#med_age.head()
```

```python
#Calculate Poverty Rate
pov_rate = Essex_df.groupby(['City'])['Poverty Rate'].mean()
#pov_rate.head()
```

```python
#Calculate Poverty Count
povcount = Essex_df.groupby(['City'])['Poverty Count'].mean()
#povcount.head()
```

```python
#Calculate Population
totpop = Essex_df['Population'].sum()
pop = Essex_df.groupby(['City'])['Population'].sum()
poppercent = Essex_df.groupby(['City'])['Population'].sum()/totpop*100
#poppercent
```

```python
#Calculate Household Income
usorigin = Essex_df.groupby(['City'])['U.S. Born'].sum()
fororigin = Essex_df.groupby(['City'])['Foreign Born'].sum()
#usorigin.head()
```

```python
# Place all of the data found into a summary DataFrame
pd.options.display.float_format = '{:,.2f}'.format
summary_table = pd.DataFrame({"Average House Hold Income": hh_income,
                              "Average Median Age": med_age,
                              "Population": pop,
                              "Percent": poppercent,
                              "Average Poverty Rate": pov_rate,
                              "Average Poverty Count": povcount,
                              "US Born": usorigin,
                              "Foreign Born": fororigin,
                             })
summary_table['Population'] = summary_table['Population'].map('{:,}'.format)
summary_table['Average House Hold Income'] = summary_table['Average House Hold Income'].map('${:,.2f}'.format)
summary_table['US Born'] = summary_table['US Born'].map('{:,}'.format)
summary_table['Foreign Born'] = summary_table['Foreign Born'].map('{:,}'.format)
summary_table['Percent'] = summary_table['Percent'].map('{:,.2f}%'.format)
summary_table['Average Poverty Rate'] = summary_table['Average Poverty Rate'].map('{:,.2f}%'.format)
summary_table
```

```python
# Export file as a CSV, without the Pandas index, but with the header
summary_table.to_csv("Essex_County_Census.csv", index=False, header=True)
```
